```python
# İlk olarak kütüphaneleri içeri aktarıyoruz.

import pandas as pd
import numpy as np
```

```python
# Çalışma yapacağımız verisetini df adına tanımlıyoruz

df = pd.read_csv('spotify.csv')
df.head()
```

```python
# Veriseti hakkında bilgilere bakıyoruz. Boş veri var mı hangi sütun hangi veri tipinde tanımlanmış vs.

df.info()
```

```python
# Verisetimizde boş varmı var ise kaç adet var onu kontrol ediyoruz.
# Görüldüğü üzere artists, album name ve track name de birer adet boş veri olduğunu görüyoruz.

df.isnull().sum()
```

```python
# describe bize tüm sütunların istatiksel analizini vermektedir.

df.describe()
```

```python
# Verisetimizin ilk 5 satırına bakıyoruz.

df.head()
```

```python
# Boş olan sütunları kaldırarak veri setimizdeki verileri eşitliyoruz.

df = df.dropna()
```

```python
df.info()
```

```python
df.head()
```

```python
# sadece işlem yapacağımız sütunları alarak veriyi manüpile ediyoruz.

df = df[["track_name","track_artist","track_popularity","playlist_name","danceability","duration_ms","danceability"]]
```

```python
df.drop ('track_name', axis=1, inplace=True)
```

```python
df.head()
```

```python
# kaç farklı playlist var ve bunlar kaç şarkı içeriyor ona bakıyoruz. en çok şarkı içeren playlist indie poptimisim.

total_playlists = df["playlist_name"].value_counts().reset_index()
total_playlists

df["playlist_name"] = pd.to_numeric(df["playlist_name"], errors="coerce")

df.head()

# to_group = df.groupby(["track_artist"]).agg({"playlist_name": ['count']})

# to_sorted = to_group.sort_values(by="playlist_name", ascending = False)
```

```python
df["track_popularity"]
```

```python
# En popüler ilk 5 sanatçıyı listeliyoruz.

df['track_popularity'] = pd.to_numeric(df['track_popularity'], errors = 'coerce')

popular_artists = df.groupby('track_artist')['track_popularity'].max().reset_index()
sorted_data = popular_artists.sort_values(by='track_popularity', ascending=False)
sorted_data.head()
```

```python
# milisaniye olarak yazan şarkı sürelerini dakika cinsine çevirip yeni bir sütun ekliyoruz.

df.loc[:, "duration_ms"] = df["duration_ms"] / (1000*60)
df.head()
```

```python
# dakika cinsinden ortalamayı alıyoruz.
# ortalama şarkı süresi 3.8 dakika olarak çıkıyor

df["duration_ms"].mean()
```

```python
df['duration_ms']
```

```python
# En uzun ilk 5 şarkı ve en kısa 5 şarkı listelenmiştir.

to_int_code = pd.to_numeric(df['duration_ms'] , errors = 'coerce')

group_data = df.groupby('track_name')['duration_ms'].mean().reset_index()

sort_data = group_data.sort_values(by='duration_ms', ascending=False)

sort_data
```
